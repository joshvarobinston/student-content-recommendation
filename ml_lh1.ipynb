```python
from content.models import ContentItem
from content.services.text_processing import build_content_text
from content.services.tfidf import build_tfidf_matrix

texts = [build_content_text(c) for c in ContentItem.objects.all()]
matrix, vectorizer = build_tfidf_matrix(texts)

print(matrix.shape)
```

```python
from django.contrib.auth.models import User
from content.services.user_profile import build_user_profile_text

user = User.objects.first()
print(build_user_profile_text(user))
```

```python
profile_text = build_user_profile_text(user)
print("PROFILE TEXT:", repr(profile_text))
```

```python
from content.services.similarity import compute_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Build user vector using same vectorizer
user_text = build_user_profile_text(user)

if user_text.strip():
    user_vector = vectorizer.transform([user_text])
    scores = compute_similarity(user_vector, matrix)
    print(scores)
else:
    print("Cold start user â€” no profile yet")
```
